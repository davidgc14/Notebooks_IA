```python
import numpy as np 
import scipy as sc 
#import sklearn as sk 

import matplotlib.pyplot as plt 
```

```python
#Definimos una funcion anónima
#Vamos a usar esta función 
#np.sin((1 / 2) * x**2 - (1 / 4) * y**2 + 3) * np.cos(2 * x + 1 - np.e**y), 
#pero en función de un vector de parámetros th = (x,y)
func = lambda th: np.sin((1 / 2) * th[0]**2 - (1 / 4) * th[1]**2 + 3) * np.cos(2 * th[0] + 1 - np.e**th[1])

#podemos comprobar el funcionamiento de nuestra función mediante la siguiente orden
#func([5, 3])

#ahora vamos a dibujar el espacio. para ello consideramos el siguiente valor, que va dibujado de -2 a 2, y en el recorrido del vector vamos a considerar 100 valores (por ejemplo)
res = 100 #el 100 va a significar la resolución de la gráfica resultante
V_x = np.linspace(-2, 2, res)
V_y = np.linspace(-2, 2, res)
#es decir estamos creando los ejes de coordenadas

#ahora lo que queremos es que para cada convinación de cualquiera dos valores de V_x y V_y nos de la altura de la superficie que queremos graficar, por lo que definimos la matriz V_z
V_z = np.zeros((res, res)) #inicializamos la matriz con valores nulos. Su dimensión es res * res por ser el número de convinaciones de los dos vectores anteriores

#vamos ahora a rellenar la matriz con sus valores
for ix, x in enumerate(V_x): #enumerate devuelve cada hueco del vector con su índice (que lo guardamos en ix) y su valor (que lo guardamos en x)
    for iy, y in enumerate(V_y): #es decir, estamos diciendo que los indices son xi y xy, y que vamos a usar x e y como los valores del propio vector
        V_z[iy, ix] = func([x,y]) #ponemos primero el indice en y para que se nos represente en horizontal dichos valores, y no se nos quede la matriz traspuesta
```

```python
#plt.contour? el interrogante sirve para saber informació de la función y como implementarla
plt.contour(V_x, V_y, V_z, 100) #último parámetro es para aumentar la resolución
plt.show() 
```
